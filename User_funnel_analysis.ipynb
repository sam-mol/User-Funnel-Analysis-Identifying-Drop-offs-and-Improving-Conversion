## User Funnel Analysis: Identifying Drop-offs and Improving Conversion

Here we are exploring existing data to understand user behavior.
We ask:

Where do users drop off?
Which step is a bottleneck?
Which segment struggles most?

```python
# Dataset generation

import numpy as np
import pandas as pd

np.random.seed(42)

n_users = 10000

df = pd.DataFrame({
    'user_id': np.arange(n_users),
    'visited': np.ones(n_users)
})

# Funnel probabilities
signup_prob = 0.6
activation_prob = 0.5
purchase_prob = 0.3

df['signed_up'] = df['visited'] * np.random.binomial(1, signup_prob, n_users)
df['activated'] = df['signed_up'] * np.random.binomial(1, activation_prob, n_users)
df['purchased'] = df['activated'] * np.random.binomial(1, purchase_prob, n_users)

# Add segmentation
df['device'] = np.random.choice(['mobile', 'desktop'], size=n_users, p=[0.7, 0.3])

df.head()
```

```python
# Compute conversion rates

funnel = {
    "visit": df['visited'].sum(),
    "signup": df['signed_up'].sum(),
    "activation": df['activated'].sum(),
    "purchase": df['purchased'].sum()
}

for step, count in funnel.items():
    print(step, int(count))
```

```python
# Drop-off analysis

signup_rate = df['signed_up'].mean()
activation_rate = df['activated'].sum() / df['signed_up'].sum()
purchase_rate = df['purchased'].sum() / df['activated'].sum()

print(f"Signup rate: {signup_rate:.3f}")
print(f"Activation rate: {activation_rate:.3f}")
print(f"Purchase rate: {purchase_rate:.3f}")
```

```python
# Segmentation analysis

df.groupby('device')[['signed_up', 'activated', 'purchased']].mean()
```

### Insights

Purchase step has the largest drop-off overall. 

Desktop users convert less at all stages.

### Recommendations

Simplify signup and activation steps.

Reduce friction in purchasing step by remove all unnecessary steps to make the purchase as quick as possible.

Improve product descriptions and address potential customer questions to reduce purchase hesitation. 

Use targetted adds and purchase discounts.
